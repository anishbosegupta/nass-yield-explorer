```python

import pandas as pd

df_raw = pd.read_csv("corn_yield.csv")

# Keep only what we need
df = df_raw[["Year", "State", "Value"]].copy()
df.columns = ["year", "state", "yield_bu_acre"]
df["state"] = df["state"].str.title()  # "IOWA" → "Iowa"
df = df.sort_values(["state", "year"]).reset_index(drop=True)

print(df.shape)
display(df.head(10))
```

```python
df.to_csv("nass_corn_clean.csv", index=False)
```
